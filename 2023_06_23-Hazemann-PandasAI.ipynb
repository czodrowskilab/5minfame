```python
import pandas as pd
from pandasai import PandasAI

#Instantiate a LLM
from pandasai.llm.openai import OpenAI
llm = OpenAI(api_token="replace_with_your_key")
pandas_ai = PandasAI(llm)
```

```python
df = pd.read_csv('~/data/hERG_ChEMBL_simple.csv')
```

```python
df
```

```python
pandas_ai.run(df, prompt='How many rows in df?')
```

```python
pandas_ai.run(df, prompt='Drop Molecule Name column')
```

```python
df
```

```python
pandas_ai.run(df, prompt='Plot the distribution of rows over Document Year')
```

```python
pandas_ai.run(df, prompt='Label in a new column with herg_IC50_nM below 5000 as risky, herg_IC50_nM above 5000 but below 10000 as moderate_risk and herg_IC50_nM above 10000 as no_risk')
```

```python
pandas_ai.run(df, prompt='Plot the distribution of Risk Level and color the bars with different colors')
```

```python
pandas_ai.run(df, prompt='Convert hERG_IC50_nM to pIC50')
```
